Task 1 : Fabry-Perot interference: analytical calculation

```python
import numpy as np
#頻率相關參數
Nf = 200
fmin = 1
fmax = 5
fre = np.linspace(fmin, fmax, Nf)
omega = 2 * np.pi * fre
```

```python
#晶體厚度
d_list = [0.167,0.056]
#材料介電常數
epsi_list = [9,80]
```

```python
#材料相位差
delta_list = []
for epsi, d in zip(epsi_list, d_list):
    delta = 2*omega * d * np.sqrt(epsi)
    delta_list.append(delta)
```

```python
#反射係數
r_list = []
for epsi in epsi_list:
    r = (np.sqrt(epsi) - 1)/( np.sqrt(epsi) + 1 )
    r_list.append(r)
#透射率
T_ana_list = []
for r, delta in zip(r_list, delta_list):
    R = np.abs(r**2)
    F = 4 * R/(1 - R)**2
    T_ana = 1 / (1 + F * np.sin(delta/2)**2)
    T_ana_list.append(T_ana)
```

```python
#頻率-透射率 圖形
from matplotlib import pyplot as plt 
plt.figure(dpi=100)
line_color = ["blue", "green"]
line_style = ["--", "-"]
for T_ana, line_color, line_style in zip(T_ana_list, line_color, line_style):
    plt.plot(fre,T_ana,color=line_color, linestyle=line_style)
plt.xlabel('$\omega_0$/$2\pi$',fontsize = 16)
plt.ylabel('T($\omega_0$)',fontsize = 16)
plt.xlim([fmin,fmax])
plt.ylim([0, 1])
plt.grid()
plt.show()
```
