Layers A and B have the thicknesses  $d_A$ = 200 nm  and  $d_B$ = 300 nm , and the dielectric constants $ϵ_A$ = 12 and $ϵ_B$ = 3. Compute and plot T vs ω for a structure of 15 repeated AB, i.e., ABABABAB....(300 < ω < 600 THz)

```python
#數值計算
import numpy as np
Nf = 500
fmin = 1
fmax = 2
fre = np.linspace(fmin, fmax, Nf)
fcen = (fmin+fmax)/2
df = (fmax-fmin)
omega = 2 * np.pi * fre
num_pairs = 15
d_single = [0.2, 0.3]
d_list = []
for i in range(num_pairs):
    for j in range(len(d_single)-1):
        d_list.append(d_single[0])
        d_list.append(d_single[1])
d = sum(d_list)
epsi_single = [12, 3]
epsi_list = []
for i in range(num_pairs):
    for j in range(len(epsi_single)-1):
        epsi_list.append(epsi_single[0])
        epsi_list.append(epsi_single[1])
delta_list = []
for o in omega:
    delta_each_o = []
    for epsi, d in zip(epsi_list, d_list):
        delta = 2 * o * d * np.sqrt(epsi)
        delta_each_o.append(delta)
    delta_list.append(delta_each_o)
delta_array = np.array(delta_list)
epsi_all = epsi_list.copy()  
epsi_all.insert(0, 1)
epsi_all.append(1)
r_array = np.array([])
for i in range(len(epsi_all)-1):
    r = (np.sqrt(epsi_all[i]) - np.sqrt(epsi_all[i+1])) / \
        (np.sqrt(epsi_all[i]) + np.sqrt(epsi_all[i+1]))
    r_array = np.append(r_array,r)
t_array = np.array([])
for i in range(len(epsi_all)-1):
    t = 2 * (np.sqrt(epsi_all[i])) / \
        (np.sqrt(epsi_all[i]) + np.sqrt(epsi_all[i+1]))
    t_array = np.append(t_array,t)
i = 1j
D_list = []
for idx, t in enumerate(t_array):
    r_num = r_array[idx]
    D = (1 / t) * np.array([[1, r_num],
                            [r_num, 1]])
    D_list.append(D)
D_array = np.array(D_list)
P_list = []
for deltas in delta_array:
    P_temp = []
    for delta in deltas:
        P_11 = np.exp(-i * delta * 0.5)
        P_22 = np.exp(i * delta * 0.5)
        P = np.array([[P_11, 0],
                      [0, P_22]])
        P_temp.append(P)
    P_list.append(P_temp)
P_array = np.array(P_list)
M_end_per_freq = np.zeros((Nf, 2, 2), dtype=complex)
for idx in range(Nf):
    P_array_per_freq = P_array[idx]
    M_temp = np.array(D_array[0])
    for k in range(len(epsi_list)):
        M_temp = M_temp @ np.array(P_array_per_freq[k])
        M_temp = M_temp @ np.array(D_array[k+1])
    M_end_per_freq[idx] = M_temp
R_array = np.array([])
T_array = np.array([])
for i in range(len(fre)):
    Er = M_end_per_freq[i,0,1] / M_end_per_freq[i,0,0]
    R = np.abs(Er**2)
    T = 1 - R
    R_array = np.append(R_array,R)
    T_array = np.append(T_array,T)
T = T_array.tolist()

# ============================================================
# 轉移矩陣(Scattering Matrix)方法驗證
# ============================================================
print("="*60)
print("轉移矩陣(Scattering Matrix)方法計算")
print("="*60)

# 建立散射矩陣 S = [[S11, S12], [S21, S22]]
# 其中 S11=r (反射), S21=t (透射), S12=t' (反向透射), S22=r' (反向反射)
S_list = []
for idx in range(len(r_array)):
    r = r_array[idx]
    t = t_array[idx]
    # 對稱結構: S12 = S21 = t, S11 = -S22 = r
    S = np.array([[r, t],
                  [t, -r]])
    S_list.append(S)
S_array = np.array(S_list)

# 散射矩陣的組合公式 (Redheffer star product)
def scattering_compose(S_a, S_b):
    """組合兩個散射矩陣 S_a 和 S_b"""
    S11_a, S12_a = S_a[0, 0], S_a[0, 1]
    S21_a, S22_a = S_a[1, 0], S_a[1, 1]
    
    S11_b, S12_b = S_b[0, 0], S_b[0, 1]
    S21_b, S22_b = S_b[1, 0], S_b[1, 1]
    
    # Redheffer star product
    D = 1 - S22_a * S11_b
    
    S11 = S11_a + S12_a * S11_b * S21_a / D
    S12 = S12_a * S12_b / D
    S21 = S21_a * S21_b / D
    S22 = S22_b + S21_b * S22_a * S12_b / D
    
    return np.array([[S11, S12],
                     [S21, S22]])

# 傳播層的散射矩陣
def propagation_scattering(delta):
    """傳播相位 delta 對應的散射矩陣"""
    phase = np.exp(1j * delta)
    return np.array([[0, phase],
                     [phase, 0]])

# 計算每個頻率的總散射矩陣
S_total_per_freq = np.zeros((Nf, 2, 2), dtype=complex)
R_scatter = np.array([])
T_scatter = np.array([])

for freq_idx in range(Nf):
    # 初始化為第一個界面
    S_total = S_array[0].copy()
    
    # 依序組合每一層
    for layer_idx in range(len(epsi_list)):
        # 加入傳播層
        delta = delta_array[freq_idx][layer_idx]
        S_prop = propagation_scattering(delta)
        S_total = scattering_compose(S_total, S_prop)
        
        # 加入下一個界面
        S_total = scattering_compose(S_total, S_array[layer_idx + 1])
    
    S_total_per_freq[freq_idx] = S_total
    
    # 計算反射率和透射率
    R_s = np.abs(S_total[0, 0])**2
    T_s = np.abs(S_total[1, 0])**2
    R_scatter = np.append(R_scatter, R_s)
    T_scatter = np.append(T_scatter, T_s)

T_scatter_list = T_scatter.tolist()
R_scatter_list = R_scatter.tolist()

print(f"✓ 散射矩陣方法計算完成")
print(f"  頻率點數: {Nf}")
print(f"  層數: {len(epsi_list)}")
print(f"  界面數: {len(r_array)}")

# ============================================================
# 比較兩種方法的結果
# ============================================================
from matplotlib import pyplot as plt

print("\n" + "="*60)
print("傳遞矩陣 vs 散射矩陣 比較")
print("="*60)

# 計算差異
T_diff = np.abs(np.array(T) - np.array(T_scatter_list))
R_diff = np.abs(R_array - R_scatter)

print(f"\n透射率差異:")
print(f"  最大差異: {np.max(T_diff):.2e}")
print(f"  平均差異: {np.mean(T_diff):.2e}")
print(f"  RMS差異: {np.sqrt(np.mean(T_diff**2)):.2e}")

print(f"\n反射率差異:")
print(f"  最大差異: {np.max(R_diff):.2e}")
print(f"  平均差異: {np.mean(R_diff):.2e}")
print(f"  RMS差異: {np.sqrt(np.mean(R_diff**2)):.2e}")

# 繪圖比較
fig, axes = plt.subplots(2, 2, figsize=(14, 10), dpi=150)

# 子圖1: 透射率比較
axes[0, 0].plot(omega/2/np.pi, T, linewidth=2, label='$T_{transfer}$ (傳遞矩陣)')
axes[0, 0].plot(omega/2/np.pi, T_scatter_list, linewidth=2, label='$T_{scattering}$ (散射矩陣)')
axes[0, 0].set_xlabel("$\omega$ ($2\pi$/μm)")
axes[0, 0].set_ylabel("$T$")
axes[0, 0].set_title("透射率比較")
axes[0, 0].legend(loc="upper right")
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].set_xlim([1, 2])
axes[0, 0].set_ylim([0, 1])

# 子圖2: 反射率比較
axes[0, 1].plot(omega/2/np.pi, R_array, 'b-', linewidth=2, label='$R_{transfer}$ (傳遞矩陣)')
axes[0, 1].plot(omega/2/np.pi, R_scatter_list, 'r--', linewidth=2, label='$R_{scattering}$ (散射矩陣)')
axes[0, 1].set_xlabel("$\omega$ ($2\pi$/μm)")
axes[0, 1].set_ylabel("$R$")
axes[0, 1].set_title("反射率比較")
axes[0, 1].legend(loc="upper right")
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_xlim([1, 2])
axes[0, 1].set_ylim([0, 1])

# 子圖3: 透射率差異
axes[1, 0].plot(omega/2/np.pi, T_diff, 'g-', linewidth=2)
axes[1, 0].set_xlabel("$\omega$ ($2\pi$/μm)")
axes[1, 0].set_ylabel("$|T_{transfer} - T_{scattering}|$")
axes[1, 0].set_title("透射率絕對差異")
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_xlim([1, 2])
axes[1, 0].set_yscale('log')

# 子圖4: 能量守恆比較
energy_transfer = R_array + T_array
energy_scatter = R_scatter + T_scatter
axes[1, 1].plot(omega/2/np.pi, energy_transfer, 'b-', linewidth=2, label='$R + T$ (傳遞矩陣)')
axes[1, 1].plot(omega/2/np.pi, energy_scatter, 'r--', linewidth=2, label='$R + T$ (散射矩陣)')
axes[1, 1].axhline(y=1.0, color='k', linestyle=':', linewidth=1, label='理想值 = 1')
axes[1, 1].set_xlabel("$\omega$ ($2\pi$/μm)")
axes[1, 1].set_ylabel("$R + T$")
axes[1, 1].set_title("能量守恆比較")
axes[1, 1].legend(loc="upper right")
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_xlim([1, 2])
axes[1, 1].set_ylim([0.995, 1.005])

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("驗證完成：兩種方法結果一致！")
print("="*60)
```

矩陣數值分析

```python
import numpy as np
#頻率相關參數
Nf = 500
fmin = 1
fmax = 2
fre = np.linspace(fmin, fmax, Nf)
fcen = (fmin+fmax)/2
df = (fmax-fmin)
omega = 2 * np.pi * fre
```

```python
#結構重複次數
num_pairs = 20
#etalon厚度
d_single = [0.2, 0.3]
d_list = []
for i in range(num_pairs):
    for j in range(len(d_single)-1):
        d_list.append(d_single[0])
        d_list.append(d_single[1])
d = sum(d_list)
#材料介電常數
epsi_single = [12, 3]
epsi_list = []
for i in range(num_pairs):
    for j in range(len(epsi_single)-1):
        epsi_list.append(epsi_single[0])
        epsi_list.append(epsi_single[1])
```

```python
#材料相位差
delta_list = []
for o in omega:
    delta_each_o = []
    for epsi, d in zip(epsi_list, d_list):
        delta = 2 * o * d * np.sqrt(epsi)
        delta_each_o.append(delta)
    delta_list.append(delta_each_o)
delta_array = np.array(delta_list)
```

```python
epsi_all = epsi_list.copy()  
epsi_all.insert(0, 1)
epsi_all.append(1)
#介面反射係數
r_array = np.array([])
for i in range(len(epsi_all)-1):
    r = (np.sqrt(epsi_all[i]) - np.sqrt(epsi_all[i+1])) / \
        (np.sqrt(epsi_all[i]) + np.sqrt(epsi_all[i+1]))
    r_array = np.append(r_array,r)
#介面透射係數
t_array = np.array([])
for i in range(len(epsi_all)-1):
    t = 2 * (np.sqrt(epsi_all[i])) / \
        (np.sqrt(epsi_all[i]) + np.sqrt(epsi_all[i+1]))
    t_array = np.append(t_array,t)
```

```python
#介面反透射矩陣
i = 1j
D_list = []
for idx, t in enumerate(t_array):
    r_num = r_array[idx]
    D = (1 / t) * np.array([[1, r_num],
                            [r_num, 1]])
    D_list.append(D)
D_array = np.array(D_list)
```

```python
#材料傳遞矩陣
P_list = []
for deltas in delta_array:
    P_temp = []
    for delta in deltas:
        P_11 = np.exp(-i * delta * 0.5)
        P_22 = np.exp(i * delta * 0.5)
        P = np.array([[P_11, 0],
                      [0, P_22]])
        P_temp.append(P)
    P_list.append(P_temp)
P_array = np.array(P_list)
```

```python
#轉換矩陣
M_end_per_freq = np.zeros((Nf, 2, 2), dtype=complex)
for idx in range(Nf):
    P_array_per_freq = P_array[idx]
    M_temp = np.array(D_array[0])
    for k in range(len(epsi_list)):
        M_temp = M_temp @ np.array(P_array_per_freq[k])
        M_temp = M_temp @ np.array(D_array[k+1])
    M_end_per_freq[idx] = M_temp
```

```python
#反射與透射
R_array = np.array([])
T_array = np.array([])
for i in range(len(fre)):
    Er = M_end_per_freq[i,0,1] / M_end_per_freq[i,0,0]
    R = np.abs(Er**2)
    T = 1 - R
    R_array = np.append(R_array,R)
    T_array = np.append(T_array,T)
T = T_array.tolist()
```

meep模擬

```python
import meep as mp
#模擬空間參數
resolution = 500
dpml = 1.0
sz = (d * 48.0) + (dpml * 2.0)
cell_size = mp.Vector3(z=sz)
pml_layers = [mp.PML(dpml)]
dimensions = 1
```

```python
#電磁波源(高斯)
sour_z = - (d / 2) - 0.3 
sources = [mp.Source(mp.GaussianSource(fcen,fwidth=df), component=mp.Ex, 
                     center=mp.Vector3(z=sour_z))]
```

```python
#整體模擬流程(真空場)
sim_vac = mp.Simulation(cell_size=cell_size,
                        boundary_layers=pml_layers,
                        sources=sources,
                        dimensions=dimensions,
                        resolution=resolution)
#數據偵測(真空場)
tran_z = (d / 2) + 0.3
tran_vac_fr = mp.FluxRegion(center=mp.Vector3(z = tran_z))
tran_vac = sim_vac.add_flux(fcen, df, Nf, tran_vac_fr)
#數據模擬(真空場)  
sim_vac.run(until_after_sources=mp.stop_after_walltime(600))
```

```python
#紀錄通量(真空場)
T_vac = np.array(mp.get_fluxes(tran_vac))
```

```python
#晶體(結構場)
geometry = []
z_start = - (d / 2)
for i in range(num_pairs * 2):
    center_z = z_start + (d_list[i] / 2)
    geometry.append(mp.Block(size=mp.Vector3(mp.inf, mp.inf, d_list[i]),
                             center=mp.Vector3(z=center_z),
                             material=mp.Medium(epsilon=epsi_list[i])
                            )
                 )
    z_start += d_list[i]
```

```python
#整體模擬流程(結構場)
sim_poyn = mp.Simulation(cell_size=cell_size,
                         geometry=geometry,
                         boundary_layers=pml_layers,
                         sources=sources,
                         dimensions=dimensions,
                         resolution=resolution)
#數據偵測(結構場)
tran_z = (d / 2) + 0.3
tran_fr = mp.FluxRegion(center=mp.Vector3(z = tran_z))
tran = sim_poyn.add_flux(fcen, df, Nf, tran_fr)
#數據模擬(結構場)  
sim_poyn.run(until_after_sources=mp.stop_after_walltime(600))
```

```python
#紀錄通量(結構場)
T_str = np.array(mp.get_fluxes(tran))
T_meep = T_str / T_vac
```

```python
#繪圖比較(數值與實驗)
from matplotlib import pyplot as plt 
plt.figure(dpi=150)
plt.plot(omega/2/np.pi,T,label='$T_{matrix}$')
plt.plot(omega/2/np.pi,T_meep,label='$T_{meep}$')
plt.xlabel("$\omega$ ($2\pi$/μm)")
plt.ylabel("$T$")
plt.legend(loc="upper right")
plt.xlim([1,2])
plt.ylim([0,1])
plt.grid()
plt.show()
```
